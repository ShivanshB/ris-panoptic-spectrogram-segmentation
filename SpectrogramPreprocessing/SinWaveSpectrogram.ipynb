# Sin Wave Spectrograms
In this notebook I will be generating simple spectrograms based on sin waves and a little bit of noise.

```python
# basic imports
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

# Pure Signal, No Noise
I will first plot a pure signal with no noise.

```python
# time difference between sampling one point and the next
dt = 0.001

# array of times and the basic pure signal of 1 Hz
t = np.arange(0.0, 3.0, dt)
s1 = np.sin(2*np.pi*1*t)

# setting x to the signal, s1, and setting window size and sampling frequency
x = s1
NFFT = 10000   # the length of the windowing segments
Fs = int(1.0/dt)  # the sampling frequency


# creating the figure and plotting the spectrogram
plt.figure(figsize=(12,8))
ax2 = plt.subplot(211)
plt.plot(t, x)
ax1 = plt.subplot(212)
ax1.set_xlabel("Time (s)")
ax2.set_xlabel("Time (s)")
ax1.set_ylabel("Frequency (Hz)")

# creating and plotting the spectrogram
Pxx, freqs, bins, im = plt.specgram(x, NFFT=NFFT, Fs=Fs, noverlap=255)


# creating and plotting the color bar/mapping
cbar = plt.colorbar(im)
cbar.set_label("Power (dB re 1 (m/s)^2/Hz)")


# zooming in on the relevant frequencies in the plot
ax1.set_ylim((0,5))
plt.show()
```

# Two Signals, No Noise
I will now be plotting two signals with different frequencies 

```python
# time difference between sampling one point and the next
dt = 0.001

# array of times and the basic pure signal of 1 Hz
t = np.arange(0.0, 3.0, dt)
s1 = np.sin(2*np.pi*1*t)
s2 = np.sin(2*np.pi*3*t)

# setting x to the signal, s1 + s2, and setting window size and sampling frequency
x = s1 + s2
NFFT = 10000   # the length of the windowing segments
Fs = int(1.0/dt)  # the sampling frequency


# creating the figure and plotting the spectrogram
plt.figure(figsize=(12,8))
ax2 = plt.subplot(211)
plt.plot(t, x)
ax1 = plt.subplot(212)
ax1.set_xlabel("Time (s)")
ax2.set_xlabel("Time (s)")
ax1.set_ylabel("Frequency (Hz)")

# creating and plotting the spectrogram
Pxx, freqs, bins, im = plt.specgram(x, NFFT=NFFT, Fs=Fs, noverlap=255)


# creating and plotting the color bar/mapping
cbar = plt.colorbar(im)
cbar.set_label("Power (dB re 1 (m/s)^2/Hz)")


# zooming in on the relevant frequencies in the plot
ax1.set_ylim((0,5))
plt.show()
```

# Two Signals, with Noise
I will now be plotting the spectrogram of two signals, but with noise included.

```python
# time difference between sampling one point and the next
dt = 0.001

# array of times and the basic pure signal of 1 Hz
t = np.arange(0.0, 3.0, dt)
s1 = np.sin(2*np.pi*1*t)
s2 = np.sin(2*np.pi*3*t)

# creating a data array of noise
noise = 0.1*np.random.randn(len(t))

# setting x to the signal, s1 + s2 + noise, and setting window size and sampling frequency
x = s1 + s2 + noise
NFFT = 10000   # the length of the windowing segments
Fs = int(1.0/dt)  # the sampling frequency

# creating the figure and plotting the spectrogram
plt.figure(figsize=(12,8))
ax1 = plt.subplot(211)
plt.plot(t, x)
ax1 = plt.subplot(212)
ax1.set_xlabel("Time (s)")
ax2.set_xlabel("Time (s)")
ax1.set_ylabel("Frequency (Hz)")

# creating and plotting the spectrogram
Pxx, freqs, bins, im = plt.specgram(x, NFFT=NFFT, Fs=Fs, noverlap=255)


# creating and plotting the color bar/mapping
cbar = plt.colorbar(im)
cbar.set_label("Power (dB re 1 (m/s)^2/Hz)")


# zooming in on the relevant frequencies in the plot
ax1.set_ylim((0,5))
plt.show()
```
