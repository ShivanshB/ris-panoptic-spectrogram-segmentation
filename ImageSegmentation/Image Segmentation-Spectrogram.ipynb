This notebook documents my attempts at segmenting a seismic spectrogram.

## Clustering Based Image Segmentation

This will just be a simple implementation of k-means on an image to segment it.

```python
pic = plt.imread(filepath + 'spectrogram.png')  # dividing by 255 to bring the pixel values between 0 and 1
print(pic.shape)
plt.figure(figsize=(30,80))
plt.imshow(pic)
```

```python
pic_n = pic.reshape(pic.shape[0]*pic.shape[1], pic.shape[2])
pic_n.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, random_state = 3).fit(pic_n)
clustered = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = clustered.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.figure(figsize=(30,80))
plt.imshow(cluster_pic)
```
