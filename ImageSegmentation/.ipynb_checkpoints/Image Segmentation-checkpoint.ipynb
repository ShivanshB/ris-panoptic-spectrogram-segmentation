## Threshold Segmentation

We will first work to divide the image into two regions, namely the background and the foreground by choosing one threshold value, $n$. Every pixel with a brightness value $p >= n$ will be marked black, and $p < n$ will be turned white.

First, we will import some libraries.

```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

%matplotlib inline

from scipy import ndimage
```

```python
filepath = os.getcwd() + '/ImageSegmentation/'

image = plt.imread(filepath + 'original.jpeg')

print(image.shape)
plt.imshow(image)
```

```python
gray = rgb2gray(image)
plt.imshow(gray, cmap='gray')
```

We will now iterate through the pixels of the image and manually turn each one black or white. First, we must compute the mean of the pixel values.

```python
gray_r = gray.reshape(gray.shape[0] * gray.shape[1])

for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 1
    else:
        gray_r[i] = 0

gray = gray_r.reshape(gray.shape[0], gray.shape[1])
plt.imshow(gray, cmap = 'gray')
```

We can do further segmentation by having more threshold values as follows.

```python
gray = rgb2gray(image)
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 3
    elif gray_r[i] > 0.5:
        gray_r[i] = 2
    elif gray_r[i] > 0.25:
        gray_r[i] = 1
    else:
        gray_r[i] = 0
gray = gray_r.reshape(gray.shape[0],gray.shape[1])
plt.imshow(gray, cmap='gray')
```

## Edge Detection Through Convolution by Kernels
We will be using kernels and convolutions on an image to perform basic feature extraction, in this case edges.

```python
grid = plt.imread(filepath + 'grid.jpeg')
plt.imshow(grid)
```

```python
grid_g = rgb2gray(grid)
plt.imshow(grid_g, cmap = 'gray')
```

we will now define the kernels of the filters.

```python
horizontal = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
vertical = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
```

We will now convolve this filter over the image.

```python
out_h = ndimage.convolve(grid_g, horizontal, mode = 'reflect')
out_v = ndimage.convolve(grid_g, vertical, mode = 'reflect')


fig = plt.figure(figsize = (5,7))
plt.subplot(211)
plt.imshow(out_h, cmap = 'gray')

plt.subplot(212)
plt.imshow(out_v, cmap = 'gray')
```

The laplace filter can detect both horizontal and vertical edges.

```python
laplace = np.array([[1,1,1],[1,-8,1],[1,1,1]])
out_l = ndimage.convolve(grid_g, laplace, mode = 'reflect')
plt.imshow(out_l, cmap = 'gray')
```

## Clustering Based Image Segmentation

This will just be a simple implementation of k-means on an image to segment it.

```python
pic = plt.imread(filepath + 'original.jpeg')/255  # dividing by 255 to bring the pixel values between 0 and 1
print(pic.shape)
plt.imshow(pic)
```

```python
pic_n = pic.reshape(pic.shape[0]*pic.shape[1], pic.shape[2])
pic_n.shape
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 6, random_state = 3).fit(pic_n)
clustered = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = clustered.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.imshow(cluster_pic)
```
